```python
import pandas as pd
```

```python
df_anno_droplet = pd.read_csv('data/tabula_muris/annotations_droplet.csv', index_col=0)
df_anno_droplet.head()
```

```python
df_neighbors = pd.read_csv('scdata/neighbors_tabula_muris.tsv', sep='\t', header=None, index_col=0)
df_neighbors.columns = ['neighbor_' + str(i) for i in range(df_neighbors.shape[1])]
df_neighbors.index.name = 'cell'
df_neighbors.head()
```

```python
# 何故かindexがおかしな細胞が一部存在しているらしい。
def search_cell_tis(c):
    try:
        return df_anno_droplet.loc[c].tissue + ' : ' + df_anno_droplet.loc[c].cell_ontology_class
    except:
        return ""
```

```python
search_cell_tis('10X_P7_10_CTGTGCTGTTACGTCA')
```

```python
%%time

df_neighbors_anno = df_neighbors.applymap(search_cell_tis)
```

```python
df_neighbors_anno.to_csv('scdata/neighbors_tabula_muris.10x.anno.tsv')
```
